```python
import os
os.environ["JAVA_HOME"] = "/usr/local/opt/openjdk@17"
os.environ["PATH"] = os.environ["JAVA_HOME"] + "/bin:" + os.environ["PATH"]
```

```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.context import SparkContext
```

```python
credentials_location = "../../credentials/gcp_creds.json"

spark = (
    SparkSession.builder
        .master("local[*]")
        .appName("zoomcamp")
        .config("spark.jars", "./lib/gcs-connector-hadoop3-2.2.15.jar")
        
        # 👇 THESE TWO LINES ARE REQUIRED
        .config("spark.hadoop.fs.gs.impl",
                "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFileSystem")
        .config("spark.hadoop.fs.AbstractFileSystem.gs.impl",
                "com.google.cloud.hadoop.fs.gcs.GoogleHadoopFS")

        .config("spark.hadoop.google.cloud.auth.service.account.enable", "true")
        .config("spark.hadoop.google.cloud.auth.service.account.json.keyfile",
                credentials_location)
        .getOrCreate()
)
```

```python
df_green = spark.read.parquet('gs://de-zoomcamp-484622-spark-ny-taxi/pq/green/*/*/*')
```
